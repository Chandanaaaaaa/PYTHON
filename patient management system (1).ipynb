```python
def menu():
    patients = {}
    while True:
        print("\n1. Add/View Patient Details")
        print("2. Schedule Appointment")
        print("3. View Previous Visits")
        print("4. Add Visit Record")
        print("5. Exit")
        
        choice = input("Enter your choice (1-5): ")
        
        if choice == "1":
            manage_patient_details(patients)
        elif choice == "2":
            schedule_appointment(patients)
        elif choice == "3":
            view_previous_visits(patients)
        elif choice == "4":
            add_visit_record(patients)
        elif choice == "5":
            print("Exiting program. Goodbye!")
            break
        else:
            print("Invalid choice. Please enter a number between 1 and 5.")

def manage_patient_details(patients):
    patient_name = input("Enter patient name: ")
    
    if patient_name in patients:
        print("\nPatient Details:")
        print("Name: ", patients[patient_name]["name"])
        print("Appointment Date: ", patients[patient_name]["appointment_date"])
    else:
        print("\nPatient not found. Adding new patient.")
        patients[patient_name] = {"name": patient_name, "appointment_date": None, "visits": []}
    
    # Update or add appointment date
    appointment_date = input("Enter appointment date (YYYY-MM-DD) or press Enter to skip: ")
    if appointment_date:
        patients[patient_name]["appointment_date"] = appointment_date

def schedule_appointment(patients):
    patient_name = input("Enter patient name for scheduling appointment: ")
    
    if patient_name in patients:
        appointment_date = input("Enter new appointment date (YYYY-MM-DD): ")
        patients[patient_name]["appointment_date"] = appointment_date
        print("Appointment scheduled successfully.")
    else:
        print("Patient not found. Please add patient details first.")

def add_visit_record(patients):
    patient_name = input("Enter patient name to add visit record: ")
    
    if patient_name in patients:
        visit_date = input("Enter visit date (YYYY-MM-DD): ")
        procedures = input("Enter procedures performed (comma separated): ")
        
        visit = {
            "date": visit_date,
            "procedures": procedures
        }
        
        patients[patient_name]["visits"].append(visit)
        print("Visit record added successfully.")
    else:
        print("Patient not found. Please add patient details first.")

def view_previous_visits(patients):
    patient_name = input("Enter patient name to view previous visits: ")
    
    if patient_name in patients and patients[patient_name]["visits"]:
        print("\nPrevious Visits:")
        for visit in patients[patient_name]["visits"]:
            print("Date:", visit["date"])
            print("Procedures:", visit["procedures"])
            print("---------------")
    elif patient_name in patients:
        print("No previous visits for this patient.")
    else:
        print("Patient not found. Please add patient details first.")

# Run the code
menu()
```
